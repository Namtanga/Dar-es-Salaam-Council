```python
#Read the file internatonal debts.csv
internationaldebts=pd.read_csv('international_debt.csv')
```

```python
newinternational_debts=pd.DataFrame()
newinternational_debts['Country_Name']= internationaldebts['country_name']
newinternational_debts['Debts']= internationaldebts['debt']

```

```python

# Assuming newinternational_debts is your DataFrame
country_sizes = newinternational_debts.groupby('Country_Name').size()

# Plotting
plots.figure(figsize=(24, 12))
country_sizes.sort_values(ascending=False).plot(kind='line', color='skyblue')

plots.title('Number of Debts by Country')
plots.xlabel('Country Name')
plots.ylabel('Number of Debts')
plots.xticks(rotation=45, ha='right')
plots.tight_layout()  # Adjust layout to make room for x labels
plots.show()
```
